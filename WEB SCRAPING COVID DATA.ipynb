```python
import pandas as pd
import urllib.request
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
link = 'https://www.mohfw.gov.in/'
page = urllib.request.urlopen(link)
soup = BeautifulSoup(page)

#URLs can be fetched using the Python module urllib.request.
```

```python

page = "https://www.worldometers.info/coronavirus/"
data = requests.get(page)
soup = BeautifulSoup(data.content, "html.parser")
```

```python
print (soup)
```

```python
print("page Title is \n",soup.title.string)
```

###### Fetch  all  links in the page 

```python
all_links = soup.find_all("a")
for link in all_links:
    linking = link.get('href')
```

```python
table=soup.find_all('table',class_='main_table_countries_div')


#main_table_countries_div
```

>- #### STEPS

- **step- 1** - Get the table head from this website
- **step- 2** - In Tablehead **theadget** all the table rows
- **step- 3** - Get the table body content
- **step- 4** - Get all the rows in table body



```python
#get required content from the table

# get the table head



thead = soup.table.find('thead')
print("Get the table head from this website \n\n", thead)
```

```python
# get all the rows in the table head   #thaed
head = thead.find_all('tr')
print(head)
```

```python
# get the table body content

tbody = soup.table.find('tbody')
print(tbody)
```

```python
# get all the rows in table body

body = tbody.find_all('tr')
print(body)
```

### Table content store into a list

```python
# In table ,column title
head_column = []

# table contents, table rows 
rowvalues = []

 ### ALl rows data
    
for tr in head:
    td = tr.find_all(['th', 'td'])
    row = [i.text for i in td]
    head_column.append(row)

```

```python
print(head_column)
```

```python

# Now,body and append each row to body
for tr in body:
    td = tr.find_all(['th', 'td'])
    row = [i.text for i in td]
    rowvalues.append(row)
```

```python
print(rowvalues)
```

#### Create a DataFrame, which sores the content of data


```python
df = pd.DataFrame(rowvalues,columns=head_column[0])
```
